```python
import gradio as gr  #to construct web components
import tensorflow as tf#perform operation on neural networks
import numpy as np #numpy is numeric python
```

```python
def new_func():
    (x_train,y_train),(x_test,y_test) = tf.keras.datasets.mnist.load_data()

new_func()
# whatever we use a dataset , we divide it into 2 parts
#independent are input and dependent are output
#independent = x and dependent = y (common practice)
#kuch data traning ke liye kuch part testing ke liye 
#minist free dataset from tensorflow , small images of handwritten digits and we
#we are using it into train and test.
```

```python
x_train[0] # all are pixel format
# picels of a single image
# o stands for black color and 1 stands for white color
#max range is 0 to 253
```

```python
import matplotlib.pyplot as plt
plt.imshow(x_train[7],cmap='gray') # to show the image we use imshow
plt.show() #imshow convert the pixels into images
```

```python
#if data has high std , the data doesn't train very well...
#some are 0 and some are 253 so high std so not good training of data
#to deal with this thing we use normalisation.
x_train = x_train/255.0
x_test = x_test/255.0 #doing normalisation to make std close enough.
#so it won't be able to predict the good accuracy from the data
```

```python
x_train[0]
#so we can see that all values does not have a high standard deviation
#so training of data will be good and efficient.
```

```python
x_train = x_train.reshape(-1,28,28,1)
x_test = x_test.reshape(-1,28,28,1)
#reshaping to 28 rows and 28 columns
#-1 figures out kitni image hai in the set and 1 acts as another 
#dimension to image
#converting 2d to 3d putting 1 at the last
```

```python
y_test[0]
```

```python
y_train=tf.keras.utils.to_categorical(y_train,num_classes=10)
y_test=tf.keras.utils.to_categorical(y_test,num_classes=10)
#we divided the images into classes from the info inside y
#from 0 to 9 range so 10 uunique values.......
```

```python
model=tf.keras.Sequential()
model.add(tf.keras.layers.Conv2D(32,(3,3),activation = 'relu',input_shape=(28,28,1)))
#input_shape = (28,28,1)
model.add(tf.keras.layers.MaxPooling2D(2,2))
model.add(tf.keras.layers.Flatten())
model.add(tf.keras.layers.Dense(128,activation='relu'))
model.add(tf.keras.layers.Dense(10,activation='softmax'))
```

1st linme - squential creates a container in which we built the nrural network

2. creating layers in covo 2d ke format me , 1st hidden layers, 
    32 filter layers 3 by 3 ke...... input shape 28*28 ka hai , image ka shape mtlb
3. next input shape use karke input de diya hai apan ne upar wali line me 
4. we perform maxpooling to convert to 2*2 
5. then we flatten to put input to the input layer in 128 ke format
6. then we passes 10 classes or 10 output to see kitne unique items ki kitni probability hai

```python
model.compile(optimizer='adam',loss='categorical_crossentropy',
            metrics=['accuracy'])

#compile the model , kitne error aarha hai , input ouput ke beech ka diff
#optimizer help karkta hai adam nam ka to reduce the error. for the high
#efficiency
```

```python
model.summary()
#total summary of the model
```

```python
model.fit(x_train,y_train,epochs=10,batch_size=128,validation_split=0.2)
#running the algo how much times
#validation split = kitne percent data prr perform karna hgaui
#batch size 128 data from the whole batch 
```

```python
accuracy = model.evaluate(x_test,y_test)
accuracy
```

```python
def sketch_recognition(img):
    x=model.predict(img.reshape(1,28,28)).argmax(axis=1)
    return x

''''image ko reshape karkke normal 1d me conver karke , 
args =  ek image diya hai , 1 diya hai 2 draw kiya , conver it into shape , toh args jo hai , wo 
2 brackets bnae the toh usme se sabse high value leta hai aur output me match karta hai , fir display karta hai 
wo method use karke ki kaunsa number sahi hai.
'''
```
